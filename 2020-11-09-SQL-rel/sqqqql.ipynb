title VARCHAR(63)
,rating VARCHAR(8)
,ratingLevel VARCHAR(125)
,ratingDescription INT(10)
,release_year INT(4)
,user_rating_score FLOAT(4)
,user_rating_size INT(4)
);

drop table if exists netflix_title; 
create table netflix_title(
show_id INT(10) NOT NULL
,type	VARCHAR(7) NOT NULL
,title	VARCHAR(104) NOT NULL
,director VARCHAR(208) NOT NULL
,cast	VARCHAR(771) NOT NULL
,country	VARCHAR(123) NOT NULL
,date_added	TIMESTAMP(6) NOT NULL
,release_year	INT(4) NOT NULL
,rating VARCHAR(8) NOT NULL
, duration	VARCHAR(10) NOT NULL
,listed_in	VARCHAR(79) NOT NULL
,description	VARCHAR(248) NOT NULL
);
